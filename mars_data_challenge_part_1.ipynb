```python
#setting up the url  
url = 'https://redplanetscience.com/'
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
#printing out the browser to review the tags
print(soup.prettify())
```

```python
#going to use the div tag and the list_text class
```

```python
#printing out the part from the website with the date, title, and summary 
#for each news article
newsInfo = soup.find_all("div", class_="list_text")
print(newsInfo)
```

```python
#still using the div tag
#for article title we need 'content_title'
#for article summary we need 'article_teaser_body'
```

```python
# newsInfo = soup.find_all("div", class_="list_text")

#scraping article title
contentTitle = soup.find_all("div", class_="content_title")
contentTitle
```

```python
# newsInfo = soup.find_all("div", class_="list_text")
#scraping article teaser
articleTeaser = soup.find_all("div", class_="article_teaser_body")
articleTeaser
```

```python
len(articleTeaser)
len(contentTitle)
len(newsInfo)
```

```python
#initializing a list for the dictionary in the for loop
newsList = []

#for loop for printing out the the article titles
#news is a number values that iterates through the 
#contentTitle and articleTeaser lists
for news in range(len(newsInfo)):
    
    #printing out the titles and teasers
    print("--------------------------")
    print(contentTitle[news].text + ":")
    print(articleTeaser[news].text)
    print(" ")
    
    #pairing the current for loop values in to a dictionary
    newsDict = {
        'title' : contentTitle[news].text,
        'preview' : articleTeaser[news].text
    }
    
    #appending the current dictionary values to the list
    newsList.append(newsDict)
```
